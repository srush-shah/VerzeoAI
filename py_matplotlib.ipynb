```python
import matplotlib.pyplot as plt
```

```python
matches = [1,2,3,4]
runs = [40,60,35,79]
```

```python
plt.scatter(matches,runs)
```

```python
plt.plot(matches,runs)
```

```python
plt.plot(matches,runs)
plt.show() #This makes sure that only the plot is shown and no returned statement is printed along with it.
```

```python
#Another way of only showing the plot is to store the return value of the call in a variable
xyz = plt.plot(matches,runs)
```

```python
plt.plot(matches,runs)
plt.xlabel("Matches") # To label x-axis on the plot
plt.ylabel("Runs Scored By Kohli") # To label y-axis on the plot
plt.title("Performance in IPL 2021") # To add title of the plot
plt.show()
```

```python
plt.plot(matches,runs)
plt.xlabel("Matches") # To label x-axis on the plot
plt.ylabel("Runs Scored By Kohli") # To label y-axis on the plot
plt.title("Performance in IPL 2021") # To add title of the plot
plt.xticks([1,2,3,4]) # Define ticks on x-axis
plt.yticks([25,60,80,100]) # Define ticks on y-axis
plt.show()
```
