## 2 and 3: loc and iloc

```python
df
```

syntax for loc - dataframeReference.loc[row labels, column labels]

syntax for iloc - dataframeReference.iloc[row indices, column indices]

```python
df.loc[["c","g","e"],["cars_per_cap","drives_right"]]
```

```python
df.iloc[[2,6,4],[2,0]]
```

```python
df.loc["nation":"g","cars_per_cap":"Nation"]
```

```python
df.iloc[1: , 1:3]
```

# Subset based on condition

1. Subset your column of interest
2. Convert the output to boolean values, using some comparison operators
3. Use the following syntax - dataframeReference[..code uptil step 2..]

```python
df["cars_per_cap"]>300 # Step 1 and 2
```

```python
df[df["cars_per_cap"]>300] # Step 3
```

```python
df[(df["cars_per_cap"]<200) & (df["Nation"] == "Egypt")] # Can't use the keyword 'and' in data frames
```

```python
df[(df["cars_per_cap"]<200) | (df["Nation"] == "Egypt")] # Can't use the keyword 'or' in data frames
```
